```python
from langchain_ollama import ChatOllama
from langchain_core.messages import HumanMessage

```

```python

# Initialize the model
# "qwen2.5-coder:7b" is the model name you used in 'ollama pull'
llm = ChatOllama(
    model="qwen2.5-coder:7b",
    temperature=0,  # Set to 0 for consistent coding results
)
```

```python

# Define your prompt
message = HumanMessage(content="Write a Python function to find the factorial of a number.")
```

```python

# Invoke the model
response = llm.invoke([message])

# Print the code output
print(response.content)
```
